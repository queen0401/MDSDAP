#### ddg analysis with single-point variants and multi-point variants

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../middlefile/final_table.tsv', sep='\t')
data.head()
```

```python
ddg_data = data[data['data source'].isin(['aimms','gdsc','ret'])]
print(len(ddg_data))
```

```python
single_data = ddg_data[ddg_data.variant.str.match('^[A-Z][0-9]*[A-Z]$')]
print(single_data.shape)
single_data.head()
```

```python
multi_data = pd.concat([ddg_data, single_data],ignore_index=True)
multi_data.drop_duplicates(subset=['id'],keep=False,inplace=True)
print(multi_data.shape)
multi_data.head()
```

```python
single_ddg_list = single_data['ddg'].tolist()
print(len(single_ddg_list))
multi_ddg_list = multi_data['ddg'].tolist()
print(len(multi_ddg_list))
```

```python
import matplotlib.pyplot as plt
from scipy import stats

single = np.array(single_ddg_list)
multi = np.array(multi_ddg_list)
print(stats.mannwhitneyu(single,multi,alternative='two-sided'))
```

```python
fig, ax = plt.subplots(figsize=(4,4))
box_plot = ax.boxplot((single, multi,), labels=['single-point variant','multi-point variant'], 
            patch_artist=True,boxprops=dict(facecolor='#CDC5BF'),
            medianprops=dict(color='#CDC5BF',linewidth=0.1,linestyle='--'),
            showmeans=True,meanline=True,showfliers=False,
            meanprops=dict(color='#EE3B3B', linestyle='-',linewidth=2))

colors = ['pink', 'blue']
ax.yaxis.grid(True)
plt.ylabel('ΔΔG')
plt.title('Mann Whitney Test P=0.84')
plt.savefig('comparision.png',dpi=300)
plt.show()
```
